```python
import numpy as np
from scipy.integrate import solve_ivp
import pysindy as ps
from scipy.signal import savgol_filter
import matplotlib.pyplot as plt
from scipy.io import loadmat
from sklearn.metrics import mean_squared_error
import random
from scipy.optimize import minimize
import cvxpy as cp
```

```python
def lorenz(t, X, sigma=10, beta=8/3, rho=28):
    x, y, z = X
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - beta * z
    return [dx, dy, dz]

def add_noise(data, percentage=0):
    # additive white gaussian noise
    noise_std = np.std(data) * (percentage / 100.0)
    noise = np.random.normal(0, noise_std, data.shape)
    return data + noise
```

```python
from mpl_toolkits.mplot3d import Axes3D

# lorenz is true system

def lorenz_no_filter(t, X):
    x, y, z = X
    dx = -10.206 * x + 10.117 * y
    dy = 25.935 * x - 0.452 * y - 0.943 * x * z
    dz = -2.676 * z + 1.01 * x * y
    return [dx, dy, dz]

def learned_savgol(t, X):
    x, y, z = X
    dx = -10.182 * x + 10.09 * y
    dy = 26.662 * x - 0.680 * y - 0.962 * x * z
    dz = -2.658 * z + 1.012 * x * y
    return [dx, dy, dz]

def learned_l1(t, X):
    x, y, z = X
    dx = -10.205 * x + 10.116 * y
    dy = 26.262 * x - 0.592 * y - 0.949 * x * z
    dz = -2.669 * z + 1.012 * x * y
    return [dx, dy, dz]

def learned_savl1(t, X):
    x, y, z = X
    dx = -10.181 * x + 10.089 * y
    dy = 26.628 * x - 0.683 * y - 0.962 * x * z
    dz = -2.658 * z + 1.012 * x * y
    return [dx, dy, dz]

def learned_l1sav(t, X):
    x, y, z = X
    dx = -10.189 * x + 10.099 * y
    dy = 26.572 * x - 0.654 * y - 0.961 * x * z
    dz = -2.66 * z + 1.012 * x * y
    return [dx, dy, dz]
```

```python
# Initialization
dt = 0.001
integrator_keywords = {}
feature_names = ['x', 'y', 'z']
# Training length and ICs
t_train = np.arange(0, 8, dt)
x0_train = [-8, 8, 27]
t_train_span = (t_train[0], t_train[-1])
# Solve forwards in time
x_train_original = solve_ivp(lorenz, t_train_span, x0_train, 
                             t_eval=t_train, **integrator_keywords).y.T
```

```python
true = solve_ivp(lorenz, t_train_span, x0_train, t_eval=t_train, **integrator_keywords).y.T
result_nofilter = solve_ivp(lorenz_no_filter, t_train_span, x0_train, t_eval=t_train, **integrator_keywords).y.T
result_savgol = solve_ivp(learned_savgol, t_train_span, x0_train, t_eval=t_train, **integrator_keywords).y.T
result_l1 = solve_ivp(learned_l1, t_train_span, x0_train, t_eval=t_train, **integrator_keywords).y.T
result_savl1 = solve_ivp(learned_savl1, t_train_span, x0_train, t_eval=t_train, **integrator_keywords).y.T
result_l1sav = solve_ivp(learned_l1sav, t_train_span, x0_train, t_eval=t_train, **integrator_keywords).y.T
```

```python
start_idx = 0  # Starting index of the subset
end_idx = 8000  # Ending index of the subset
step = 1     # Step between indices

fig = plt.figure(figsize=(11, 7))
ax = fig.add_subplot(111, projection='3d')

# True Lorenz system trajectory
ax.plot(true[start_idx:end_idx:step,0], true[start_idx:end_idx:step,1], true[start_idx:end_idx:step,2], 'c', linewidth=2, label='True Lorenz')

# Setting the title and labels
ax.set_title(f"True Lorenz System - Training Data [dt=0.001, T=8]", y=1.02)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")

# Show the plot
plt.show()
```
