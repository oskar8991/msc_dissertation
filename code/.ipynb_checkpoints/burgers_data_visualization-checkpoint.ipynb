```python
import numpy as np
from scipy.integrate import solve_ivp
import pysindy as ps
from scipy.signal import savgol_filter
import matplotlib.pyplot as plt
from scipy.io import loadmat
from sklearn.metrics import mean_squared_error
import random
from scipy.optimize import minimize
import cvxpy as cp
```

```python
data = loadmat("data/burgers.mat")
time = np.ravel(data["t"])
x = np.ravel(data["x"])
u = np.real(data["usol"])
dt = time[1] - time[0]
dx = x[1] - x[0]
```

```python
# Print shape of data
print("Shape of time:", time.shape)
print("Shape of x:", x.shape)
print("Shape of u:", u.shape)

plt.figure(figsize=(16, 4))
plt.subplot(1, 3, 1)
c = plt.pcolormesh(time, x, u)
plt.xlabel("Time (t)", fontsize=10)
plt.ylabel("Spatial Coordinate (x)", fontsize=10)
plt.title("Burgers' Training Data - $u(x, t)$ of shape (256, 101)", fontsize=11, y=1.02)
plt.colorbar(c, label="Magnitude or Value")
plt.tight_layout()  # Adjusts the layout so that there's no overlap between the colorbar and other elements
plt.show()
```
